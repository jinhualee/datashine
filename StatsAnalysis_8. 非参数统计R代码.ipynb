#### 3. $r\times s$列联表和$\chi^2$独立性检验

```r
library(vcd)
dataO=c(98,38,289)
dataA=c(67,41,262)
dataB=c(13,8,57)
dataAB=c(18,12,30)
d1=rbind(dataO,dataA,dataB,dataAB)
d=data.frame(dataO,dataA,dataB,dataAB)
chisq.test(d1,correct=T)
```
